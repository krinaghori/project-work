--------------- Smart Expense Tracker Dataset Analysis ---------------

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class SmartExpenseTracker:

    def __init__(self, filename):
        self.filename = filename
        self.df = pd.read_csv(filename)
        
```

```python
class SmartExpenseTracker(SmartExpenseTracker):

    def load_data(self):
        self.df = pd.read_csv(self.filename)
        print("Dataset loaded successfully")
        display(self.df.head())
```

```python
class SmartExpenseTracker(SmartExpenseTracker):
    
    def dataset_information(self):
         df = self.df

         display(df.head())
         print()
         display(df.tail())

         print()
         print(df.dtypes)

         print()
         print("Shape :", df.shape)
         print()

         print("Columns :", df.columns)
         print()

         print("Information :")
         print(df.info())
         print()

         print("Statistical Summary")
         display(df.describe())
    
```

```python
class SmartExpenseTracker(SmartExpenseTracker):
    
     def dataframe_operations(self):
         df = self.df

         print(df["Payment_Method"])
         print()

         print(df[df["Budget"] > 10000])
         print()

         print(df[df["Amount"] < 10000])
         print()

         df["Date"] = pd.to_datetime(df["Date"])
         df["Date"] = df["Date"].dt.strftime("%d %B %Y")
         display(df.head())

         df["Date"] = pd.to_datetime(df["Date"])
         df["Year"] = df["Date"].dt.year
         display(df.head())

         df["Month"] = df["Date"].dt.month_name()
         display(df.head())

         df = df.sort_values(by="Balance", ascending=False)
         display(df.head())

         print(df["Amount"].value_counts())
    
```

```python
class SmartExpenseTracker(SmartExpenseTracker):
    def statistics(self):
        df = self.df

        print()
        print("Maximum Amount :", df["Amount"].max())
        print("Maximum Budget :", df["Budget"].max())
        print("Maximum Balance :", df["Balance"].max())
        print("Minimum Amount :", df["Amount"].min())
        print("Minimum Budget :", df["Budget"].min())
        print("Minimum Balance :", df["Balance"].min())
        print()

        print()
        print("Average Amount :", df["Amount"].mean())
        print("Median Amount :", df["Amount"].median())
        print("Mode of Amount :", df["Amount"].mode())
        print()

        print()
        print("Total City :", df["City"].nunique())
        print()
```

```python
class SmartExpenseTracker(SmartExpenseTracker):
    def missing_values(self):
         df = self.df

         print(df.isnull().sum())

         df["User_ID"] = df["User_ID"].fillna(df["User_ID"].mean())
         df["Amount"] = df["Amount"].fillna(df["Amount"].mean())
         df["Budget"] = df["Budget"].fillna(df["Budget"].mean())
         df["Balance"] = df["Balance"].fillna(df["Balance"].mean())

         df.tail()
    
```

```python
class SmartExpenseTracker(SmartExpenseTracker):
    def numpy_array(self):
       df = self.df

       arr = df["Merchant"].to_numpy()
       print(arr)
       print()
```

```python
class SmartExpenseTracker(SmartExpenseTracker):
    def bar_chart(self):
        df = self.df

        category_expense = df.groupby("Category")["Amount"].sum().head(10)

        plt.figure(figsize=(12,5))
        plt.bar(category_expense.index, category_expense.values, color="skyblue")
        plt.title("Total Expense by Category")
        plt.xlabel("Category")
        plt.ylabel("Amount")
        plt.show()
```

```python
class SmartExpenseTracker(SmartExpenseTracker):
    def line_chart(self):
         df = self.df

         daily = df.groupby("Date")["Amount"].sum().head(10)

         plt.figure(figsize=(12,5))
         plt.plot(daily.index, daily.values, marker="o", color="red")
         plt.title("Daily Expense Trend")
         plt.xlabel("Date")
         plt.ylabel("Expense")
         plt.show()
```

```python
class SmartExpenseTracker(SmartExpenseTracker):
    def pie_chart(self):
         df = self.df

         category = df.groupby("Category")["Amount"].sum()

         plt.figure(figsize=(7,7))
         plt.pie(category.values, labels=category.index, autopct="%1.1f%%", startangle=90)
         plt.title("Expense Distribution by Category")
         plt.show()
```

```python
class SmartExpenseTracker(SmartExpenseTracker):
    def histogram(self):
        df = self.df

        plt.figure(figsize=(12,5))
        plt.hist(df["Balance"], bins=15,color="green")
        plt.title("Balance Distribution")
        plt.xlabel("Balance")
        plt.ylabel("Frequency")
        plt.show()
    
```
